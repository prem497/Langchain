```python
import os
from dotenv import load_dotenv

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain.chains.llm import LLMChain
from langchain_core.tools import Tool

# Load API Key
load_dotenv()

google_api_key = os.getenv("GOOGLE_API_KEY")

# Step 1: Initialize Gemini LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=google_api_key,
    temperature=0
)

# Step 2: Create Prompt Template
prompt = PromptTemplate.from_template(
    "Answer clearly: {question}"
)

# Step 3: Build LLMChain
qa_chain = LLMChain(
    llm=llm,
    prompt=prompt
)

# Step 4: Wrap as Tool
qa_tool = Tool(
    name="Simple QA",
    func=qa_chain.run,
    description="A basic Gemini-powered QA tool"
)

# Step 5: Use the Tool
query = "What is LangGraph in LangChain?"
answer = qa_tool.run(query)

# Output
print("\nUser Question:", query)
print("\nTool Answer:", answer)
```
